# Descriptive Analytics and Customer Profiling

```python
# Grouping the data by 'Product Purchased' to calculate statistics for each product
grouped_data = df.groupby('Product')

# Define a function to calculate mean, median, and mode
def calculate_statistics(group):
    return pd.DataFrame({
        'Mean': group.mean(),
        'Median': group.median(),
        'Mode': group.mode().iloc[0]  # Take the first mode value (in case of multiple modes)
    })

# Calculate statistics for each continuous variable for each product
summary_stats = grouped_data[['Age', 'Usage', 'Income', 'Fitness', 'Miles','Education']].apply(calculate_statistics)

# Print the summary statistics
print(summary_stats)
```

# Two-Way Contingency Tables and Probabilities

•	Objective: Use contingency tables to compute and interpret conditional and marginal probabilities.

1.	Create Contingency Tables: Use pd.crosstab() to create tables showing relationships between categorical variables (e.g., Product Purchased vs. Gender, Marital Status, Fitness).

```python
# Create a contingency table for Product Purchased vs. Gender
contingency_gender = pd.crosstab(df['Product'], df['Gender'], margins=True, margins_name="Total")
print("\n--- Contingency Table: Product Purchased vs. Gender ---")
print(contingency_gender)

# Create a contingency table for Product Purchased vs. Marital Status
contingency_marital = pd.crosstab(df['Product'], df['MaritalStatus'], margins=True, margins_name="Total")
print("\n--- Contingency Table: Product Purchased vs. Marital Status ---")
print(contingency_marital)

# Create a contingency table for Product Purchased vs. Fitness
contingency_fitness = pd.crosstab(df['Product'], df['Fitness'], margins=True, margins_name="Total")
print("\n--- Contingency Table: Product Purchased vs. Fitness ---")
print(contingency_fitness)
```

2.	Marginal Probabilities: Calculate the marginal probabilities for each treadmill (KP281, KP481, KP781).
	Example: What percentage of customers purchased each product?


```python
# Calculate the total number of customers
total_customers = len(df)

# Calculate the count of each product purchased
product_counts = df['Product'].value_counts()

# Calculate marginal probabilities (percentages)
marginal_probabilities = (product_counts / total_customers) * 100

# Display the results
print("\n--- Marginal Probabilities of Each Treadmill ---")
for product, probability in marginal_probabilities.items():
    print(f"{product}: {probability:.2f}%")
```

3.	Conditional Probabilities: Calculate conditional probabilities like:
	Probability of a male buying a specific treadmill model.
	Probability of a customer in the highest fitness category purchasing the KP781.


```python
# Calculate conditional probabilities
# 1. Probability of a male buying each treadmill model
male_purchase_prob = df[df['Gender'] == 'Male']['Product'].value_counts(normalize=True)

# 2. Probability of a customer in the highest fitness category purchasing each treadmill model
highest_fitness = df[df['Fitness'] == 5]['Product'].value_counts(normalize=True)

# 3. Probability of a single customer buying each treadmill model
single_purchase_prob = df[df['MaritalStatus'] == 'Single']['Product'].value_counts(normalize=True)

# 4. Probability of a partnered customer buying each treadmill model
partnered_purchase_prob = df[df['MaritalStatus'] == 'Partnered']['Product'].value_counts(normalize=True)

# 5. Probability of a customer aged 30 or below purchasing each treadmill model
young_purchase_prob = df[df['Age'] <= 30]['Product'].value_counts(normalize=True)

# 6. Probability of customers with income above $100,000 purchasing each treadmill model
high_income_purchase_prob = df[df['Income'] > 100000]['Product'].value_counts(normalize=True)

# 7. Probability of a customer using the treadmill more than 3 times a week purchasing each treadmill model
high_usage_purchase_prob = df[df['Usage'] > 3]['Product'].value_counts(normalize=True)

# 8. Probability of a customer expecting to run more than 10 miles per week purchasing each treadmill model
high_miles_purchase_prob = df[df['Miles'] > 10]['Product'].value_counts(normalize=True)

# 9. Probability of customers with a fitness rating of 4 or higher purchasing each treadmill model
fitness_above_4_purchase_prob = df[df['Fitness'] >= 4]['Product'].value_counts(normalize=True)

# 10. Probability of customers aged between 31 and 50 purchasing each treadmill model
age_31_50_purchase_prob = df[(df['Age'] > 30) & (df['Age'] <= 50)]['Product'].value_counts(normalize=True)

# Display results
print("\n--- Conditional Probabilities ---")
print("\nProbability of a Male Buying Each Treadmill Model:\n", male_purchase_prob)
print("\nProbability of Customers in the Highest Fitness Category Purchasing Each Model:\n", highest_fitness)
print("\nProbability of a Single Customer Buying Each Treadmill Model:\n", single_purchase_prob)
print("\nProbability of a Partnered Customer Buying Each Treadmill Model:\n", partnered_purchase_prob)
print("\nProbability of Customers Aged 30 or Below Buying Each Treadmill Model:\n", young_purchase_prob)
print("\nProbability of Customers with Income Above $100,000 Buying Each Treadmill Model:\n", high_income_purchase_prob)
print("\nProbability of Customers Using the Treadmill More Than 3 Times a Week:\n", high_usage_purchase_prob)
print("\nProbability of Customers Expecting to Run More Than 10 Miles Per Week:\n", high_miles_purchase_prob)
print("\nProbability of Customers with a Fitness Rating of 4 or Higher:\n", fitness_above_4_purchase_prob)
print("\nProbability of Customers Aged Between 31 and 50:\n", age_31_50_purchase_prob)
```
